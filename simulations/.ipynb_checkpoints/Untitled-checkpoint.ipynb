```python
import kklmc
import numpy as np
import matplotlib.pyplot as plt

params = {
    'use_OM' : True,
    
    'beta' : 0.001,
    'gamma' : 5,
    'Tf' : 1,
    
    'U0' : 1,
    'dU' : 0.5,
    
    'dim' : 1,
    'Nq_factor' : 20,
    'Nm' : 10,
    
    'x0' : -1,
    'x1' : 0,
    
    'kappa' : 0.1,
}

params['state0'] = np.zeros((params['Nm'], params['dim']))
params['system'] = kklmc.systems.DoubleWell(params)

sim = kklmc.protocols.pCN(params)

states, accepts, actions, paths, d_paths = sim.simulate(10000)

choices = np.random.choice(len(states), 1000, replace=False)
for i in choices:
    plt.plot(sim.ts, paths[i,:,0], lw=0.2, alpha=0.5)

ts = np.linspace(0, params['Tf'], 1000)
state = states[-1,:,:]
path, d_path = kklmc.utils.eval_path(params['x0'], params['x1'], ts, params['Tf'], state)
plt.plot(ts, path)

state = states[1, :, :]
kklmc.utils.compute_action(params['system'], state, params['Nq_factor'], use_relative=True, use_OM=True)
```
